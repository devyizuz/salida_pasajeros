```python
column_rename_map = {
    "sal_codigo": "codigo_pais",
    "sal_destinoint": "nombre_pais_destino",
    "sal_periodo": "fecha_periodo_year",
    "month": "fecha_periodo_mes",
    "sal_valor": "frecuencia_pasajeros"
}
aeropuerto_df_renombrado = aeropuerto_df.rename(columns=column_rename_map)
```

```python
aeropuerto_df_renombrado.head()
```

```python
plt.figure(figsize=(18, 10))
frecuencia_paises = aeropuerto_df_renombrado['codigo_pais'].value_counts()
ax = frecuencia_paises.plot(kind='bar', color='skyblue', edgecolor='black')
umbral = 50
for p in ax.patches:
    if p.get_height() > umbral:
        ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center', xytext=(0, 10), textcoords='offset points', rotation=45)
plt.xticks(rotation=45, ha='right')
plt.title('Frecuencia de codigo_pais')
plt.xlabel('Código de País')
plt.ylabel('Frecuencia')
plt.tight_layout()
plt.show()
```

```python
total_pasajeros_por_pais = aeropuerto_df_renombrado.groupby('nombre_pais_destino')['frecuencia_pasajeros'].sum()
total_pasajeros_por_pais = total_pasajeros_por_pais.sort_values(ascending=False)
plt.figure(figsize=(18, 10))
total_pasajeros_por_pais.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Cantidad Total de Pasajeros que Viajaron a Cada País")
plt.xlabel("País de Destino")
plt.ylabel("Cantidad de Pasajeros")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
top_destinations = aeropuerto_df_renombrado.groupby('nombre_pais_destino')['frecuencia_pasajeros'].sum()

top_destinations = top_destinations.sort_values(ascending=False).head(6)

plt.figure(figsize=(12, 6))
top_destinations.plot(kind='bar', color='#FFCC33')
plt.title("Top 6 Destinos Internacionales por Pasajeros")
plt.xlabel("Destino")
plt.ylabel("Cantidad de Pasajeros")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Conclusiones de la Grafica: Top 6 Destinos Internacionales


*   Observamos que Estados Unidos es el pais al que mas viajan los pasajeros desde los Aeropuertos de Medellin




```python
aeropuerto_df_renombrado['periodo'] = pd.to_datetime(aeropuerto_df_renombrado['fecha_periodo_year'].astype(str) + '-' + aeropuerto_df_renombrado['fecha_periodo_mes'].astype(str), format='%Y-%m')
tendencias_temporales = aeropuerto_df_renombrado.groupby('periodo')['frecuencia_pasajeros'].sum().reset_index()

plt.figure(figsize=(14, 8))
plt.plot(tendencias_temporales['periodo'], tendencias_temporales['frecuencia_pasajeros'], marker='o', linestyle='-', color='#362F2A', label='Valor de Salida')

plt.fill_between(tendencias_temporales['periodo'], tendencias_temporales['frecuencia_pasajeros'], color='#8B7261', alpha=0.3)
plt.fill_between(tendencias_temporales['periodo'], 0, tendencias_temporales['frecuencia_pasajeros'], where=(tendencias_temporales['periodo'] < '2020-01-01'), color='#E08747', alpha=0.5, label='Antes de 2020')
plt.fill_between(tendencias_temporales['periodo'], 0, tendencias_temporales['frecuencia_pasajeros'], where=(tendencias_temporales['periodo'] >= '2020-01-01'), color='#51D4E0', alpha=0.5, label='Desde 2020')

plt.title('Tendencia de Salidas de Pasajeros entre los años 2.007 y 2.023')
plt.xlabel('Periodo')
plt.ylabel('Cantidad de Pasajeros')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Conclusiones Grafica Tendencia Salida de Pasajeros


*   A lo largo de los años hay una tendencia general ascendiente en el valor de la salida de pasajeros, esto sugiere un crecimiento en el numero de salidad año a año.
*   Hay una caida pronunciada y significativa en el valor de pasajereos en el año 2.020, esta caida puede corresponder al inicio de la pandemia por covid-19.

*  Hacia el año 2.021 se observa una recuperación a medida que avanza el tiempo




```python
aeropuerto_df_renombrado['periodo'] = pd.to_datetime(
    aeropuerto_df_renombrado['fecha_periodo_year'].astype(str) + '-' + aeropuerto_df_renombrado['fecha_periodo_mes'].astype(str),
    format='%Y-%m'
)

top_destinations = aeropuerto_df_renombrado.groupby('nombre_pais_destino')['frecuencia_pasajeros'].sum()
top_destinations = top_destinations.sort_values(ascending=False).head(6).index

plt.figure(figsize=(20, 12))

for i, pais in enumerate(top_destinations, 1):
    pais_data = aeropuerto_df_renombrado[aeropuerto_df_renombrado['nombre_pais_destino'] == pais].copy()
    tendencias_pais = pais_data.groupby('periodo')['frecuencia_pasajeros'].sum().reset_index()

    plt.subplot(3, 2, i)
    plt.plot(tendencias_pais['periodo'], tendencias_pais['frecuencia_pasajeros'], marker='o', linestyle='-', color='#362F2A', label=pais)
    plt.fill_between(tendencias_pais['periodo'], 0, tendencias_pais['frecuencia_pasajeros'], where=(tendencias_pais['periodo'] < '2020-01-01'), color='#E08747', alpha=0.5, label='Antes de 2020')
    plt.fill_between(tendencias_pais['periodo'], 0, tendencias_pais['frecuencia_pasajeros'], where=(tendencias_pais['periodo'] >= '2020-01-01'), color='#51D4E0', alpha=0.5, label='Desde 2020')
    plt.title(f'Tendencia de Salidas para {pais}')
    plt.xlabel('Periodo')
    plt.ylabel('Frecuencia de Pasajeros')
    plt.xticks(rotation=45)
    plt.legend()

plt.tight_layout()
plt.show()

```

### Concluisiones para las gráficas de Tendencias de Salidas:



*   Las gráficas muestran una tendencia de crecimiento similar en: Estados Unidos, Panama y España, con un incremento constante hasta el 2.019 y con una caida en el 2.020, seguido de una recuperación entre los años: 2.021 y 2.022.




```python
plt.figure(figsize=(14, 8))

for pais in top_destinations:
    pais_data = aeropuerto_df_renombrado[aeropuerto_df_renombrado['nombre_pais_destino'] == pais]
    tendencias_pais = pais_data.groupby('periodo')['frecuencia_pasajeros'].sum().reset_index()
    plt.plot(tendencias_pais['periodo'], tendencias_pais['frecuencia_pasajeros'], marker='o', linestyle='-', label=pais)

plt.title('Comparación de Tendencias de Frecuencia de Pasajeros por País')
plt.xlabel('Periodo')
plt.ylabel('Frecuencia de Pasajeros')
plt.legend(title='País')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Conclusiones para la Gráfica: Tendencia de Frecuencia de Pasajeros por País

*   Impacto de la Pandemia: Todos los países muestran una caida notable alrededor del 2.020 debido a la pandemia de COVID-19. Esta caida es mas pronunciada en Estados Unidos y Pánama.
*   Recuperación Post-Pandemia: También se observa una recuperación post-pandemia con valores que superan los niveles que se estaban observando en pre-pandemia.



```python
us_data = aeropuerto_df_renombrado[aeropuerto_df_renombrado['codigo_pais'] == 'US'].copy()

us_data['periodo'] = pd.to_datetime(
    us_data['fecha_periodo_year'].astype(str) + '-' + us_data['fecha_periodo_mes'].astype(str),
    format='%Y-%m'
)

tendencias_us = us_data.groupby('periodo')['frecuencia_pasajeros'].sum().reset_index()

plt.figure(figsize=(14, 8))
plt.plot(tendencias_us['periodo'], tendencias_us['frecuencia_pasajeros'], marker='o', linestyle='-', color='b')
plt.title('Tendencia de Frecuencia de Pasajeros para US a lo largo del tiempo')
plt.xlabel('Periodo')
plt.ylabel('Frecuencia de Pasajeros')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Conclusiones para la Gráfica: Tendencia de frecuencia de pasajeros hacia US

*   El tráfico hacia Estados Unidos muestra un crecimiento constante antes y después de la pandemia.
*   Se observan fluctuaciones Estacionales: los picos en ciertos años reflejan variaciones estacionales probablemente asociados a eventos especificos que impulsarion en tráfico aéreo.



```python
us_data = aeropuerto_df_renombrado[aeropuerto_df_renombrado['codigo_pais'] == 'US'].copy()

us_data['periodo'] = pd.to_datetime(
    us_data['fecha_periodo_year'].astype(str) + '-' + us_data['fecha_periodo_mes'].astype(str),
    format='%Y-%m'
)

us_data['año'] = us_data['periodo'].dt.year
us_data['mes'] = us_data['periodo'].dt.month

mejor_mes_por_año = us_data.groupby(['año', 'mes'])['frecuencia_pasajeros'].sum().reset_index()
mejor_mes_por_año = mejor_mes_por_año.loc[mejor_mes_por_año.groupby('año')['frecuencia_pasajeros'].idxmax()]

print(mejor_mes_por_año)

plt.figure(figsize=(14, 8))
plt.plot(mejor_mes_por_año['año'].astype(str) + '-' + mejor_mes_por_año['mes'].astype(str),
         mejor_mes_por_año['frecuencia_pasajeros'], marker='o', linestyle='-', color='b')
plt.title('Mejor mes de cada año para US en términos de Frecuencia de Pasajeros')
plt.xlabel('Año-Mes')
plt.ylabel('Frecuencia de Pasajeros')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Conclusiones para la gráfica: Mejor mes de cada año para EU en terminos de Frecuencia de Pasajeros

*   Se observan variaciones estacionales, que pueden estar relacionados por los picos que se encuentran cada año.




```python
moda_mes = mejor_mes_por_año['mes'].mode()[0]
print(f"La moda del mes es: {moda_mes}")

frecuencia_meses = mejor_mes_por_año['mes'].value_counts()

plt.figure(figsize=(10, 6))
frecuencia_meses.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Frecuencia de los mejores meses en términos de Frecuencia de Pasajeros')
plt.xlabel('Mes')
plt.ylabel('Frecuencia')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

### Conclusiones para la gráfica: Mejor Mes en frecuencia de Pasajeros


*   Podemos concluir que el mes de agosto es uno de los mejores meses para viajar hacia los EU, puede ser debido a una combinación de factores como: las vacaciones de verano, el clima favorable, eventos y festivales entre otros.



```python
peor_mes_por_año = us_data.groupby(['año', 'mes'])['frecuencia_pasajeros'].sum().reset_index()
peor_mes_por_año = peor_mes_por_año.loc[peor_mes_por_año.groupby('año')['frecuencia_pasajeros'].idxmin()]

print(peor_mes_por_año)

plt.figure(figsize=(14, 8))
plt.plot(peor_mes_por_año['año'].astype(str) + '-' + peor_mes_por_año['mes'].astype(str),
         peor_mes_por_año['frecuencia_pasajeros'], marker='o', linestyle='-', color='r')
plt.title('Peor mes de cada año para US en términos de Frecuencia de Pasajeros')
plt.xlabel('Año-Mes')
plt.ylabel('Frecuencia de Pasajeros')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Conclusiones para la gráfica: Peor mes del año para US en treminos de frecuencia de Pasajeros

*   Observamos que en los valores de salida mas bajos mes a mes, tabién hay aumento significativo en el transcurso de los años.




```python
moda_mes = peor_mes_por_año['mes'].mode()[0]
print(f"La moda del mes es: {moda_mes}")

frecuencia_meses = peor_mes_por_año['mes'].value_counts()

plt.figure(figsize=(10, 6))
frecuencia_meses.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Frecuencia de los peores meses en términos de Frecuencia de Pasajeros')
plt.xlabel('Mes')
plt.ylabel('Frecuencia')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

### Conclusiones para la gráfica: Peor mes del año para US en treminos de frecuencia de Pasajeros

*   Teniendo en cuenta la moda, podemos observar que el mes de febrero se destaca a lo largo de los años com el mes que menos pasajeros visitan este pais.
